```python
# Alternative: Load data into SQLite database (works on macOS without additional setup)

import pandas as pd
import sqlalchemy as sal

# Create SQLite database connection
engine = sal.create_engine('sqlite:///retail_orders.db')

print("Successfully connected to SQLite database!")

# Load the dataframe to SQL table
df.to_sql('orders', engine, if_exists='replace', index=False)
print("Data successfully loaded to 'orders' table!")

# Verify the data was loaded
result = pd.read_sql("SELECT COUNT(*) as row_count FROM orders", engine)
print(f"Number of rows in database: {result['row_count'].iloc[0]}")

# Show first few rows from database
sample_data = pd.read_sql("SELECT * FROM orders LIMIT 5", engine)
print("\nFirst 5 rows from database:")
print(sample_data)
```
